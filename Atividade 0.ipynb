## Questão 0

```python
import matplotlib.pyplot as pyplot
import numpy as numpy

carga = numpy.array([0.20, 0.30, 0.31, 0.50, 0.51, 1.01, 1.01, 1.00, 2.00, 2.01, 2.00, 3.01])
taxa = numpy.array([2.15, 6.80, 21.86, 49.09, 85.45, 115.97, 105.06, 92.07 , 151.81, 141.94, 132.59, 142.46])

pyplot.plot(carga, taxa, 'o')
pyplot.xlabel("Carga (g)")
pyplot.ylabel("Taxa de Disparo (Hz)")
```

No gráfico, nota-se um comportamento similar a uma curva logarítmica. Ou seja, a taxa de disparo aumenta conforme a carga, mas diminuindo a taxa de crescimento confrome a progressão da curva.

### Alternativa A

```python
import matplotlib.pyplot as pyplot
import numpy as numpy

def plotFromDataSet(amostra):
    dataset = numpy.genfromtxt('https://raw.githubusercontent.com/borismarin/sample-data/master/neuro-comp/i_v.dat')

    # assumindo que o eixo de tempo está em ms
    tempo = numpy.arange(1500) / 1000
    estimulo = dataset[amostra][0]
    
    pyplot.figure()
    
    pyplot.plot(tempo, dataset[amostra][1:])
    
    pyplot.xlabel("Tempo (ms)")
    pyplot.ylabel("Potencial de Membrana (UA)")
    pyplot.title(f"Estímulo {estimulo}")
    
plotFromDataSet(0)
plotFromDataSet(4)
plotFromDataSet(9)
```

Observando os gráficos, pode-se notar que a taxa de disparos aumenta conforme o estímulo
